Ziel: Berechnung eines fairen Versicherungsbeitrags

## Vorbereitung / Brainstorming

Mögliche Zusammenhänge:
- Area/Region: manchen Gegenden in Frankreich können risikoreicher sein (z.B. Großstädte)
- VehPower, Leistung des Kfz: bei höherer Leistung ggf. riskanteres Fahren, aber Auto ist ggf. auch teurer und es wird mehr darauf aufgepasst? Dann aber auch oft höhere Schäden/Reparaturen, wenn teurer.
- VehAge: Bei neuen Autos wird ggf. mehr aufgepasst, aber auch eher eingereicht als bei alten Autos. Ältere Autos ggf. wartungsanfälliger.
- DrivAge: Mehr Schäden bei jungen Fahrer:innen (Fahranfänger) und sehr alten Fahrer:innen 
- BonusMalus: Mehr Schäden mit Malus-Punkten
- VehBrand: Unklar welche Marken, aber manche könnten höhere Reparaturkosten benötigen
- VehGas: Unklar, welchen Zshg. es geben könnte 
- Density: ggf. mehr Unfälle & Schäden, wo mehr Einwohner sind
       
Unklar:
- In welcher Einheit ist VehPower angegeben?
- Wofür stehen Kategorien von Area und VehBrand?

Recherche:
- BonusMalus in Frankreich: 50 = bester Wert (50% Rabatt auf Ausgangsprämie), man startet bei 100

### Aufteilung in Trainings-/Testdaten

```python
# Variablen, die mit AV zusammenhängen oder die man bei Vorhersage noch nicht kennen kann, ausschließen
df_train = df_filtered.drop(['Exposure', 'ClaimAmount', 'ClaimNb'], axis=1) # axis = 1: entlang der Spaltenachse ausführen (Rows: x=0)

# Separaten Datensatz für UVs und AV erstellen
X = df_train.drop('ClaimPerYear', axis=1)
y = df_train['ClaimPerYear']  # Zielvariable
```

```python
# Aufteilung in Trainings- und Testdaten
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 33) # 20% Testdaten, random_state wie seed in R

print(f'Trainingsdaten: {X_train.shape}')
print(f'Testdaten: {X_test.shape}')
```

### Feature Engineering

```python
# Logarithmische Transformation für Density, da sehr rechtsschief
X_train['Density_log'] = np.log(X_train['Density'] + 1)
X_test['Density_log'] = np.log(X_test['Density'] + 1)
```

```python
plt.hist(X_train['Density_log'], bins = 10) # deutlich normalverteilter
plt.title('Histogramm Density_log')
plt.show() 
```

```python
# One-hot-encoding für nominal skalierte Variablen
# (ähnlich Dummy-Kodierung aber für alle Kategorien, nicht k-1)
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse_output=False gibt komplettes Numpy-Array zurück, nicht nur Nicht-Null-Stellen
                                                                             # handle_unknown='ignore': unbekannte Kategorien werden ignoriert, 

X_train_nominal_encoded = onehot_encoder.fit_transform(X_train[['VehBrand', 'VehGas', 'Region']]) # doppelte Klammern um Dataframe der Variablen zu haben
X_test_nominal_encoded = onehot_encoder.transform(X_test[['VehBrand', 'VehGas', 'Region']])
```

```python
# Überprüfung
print(X_train_nominal_encoded[:5])
print(X_test_nominal_encoded[:5])
```

```python
# Ordinal encoding für Area als ordinal skalierte Variable

# Kategorien anzeigen lassen
print(X_train['Area'].unique())
print(X_test['Area'].unique())

# Anführungszeichen entfernen, da sonst Probleme
X_train['Area'] = X_train['Area'].str.replace("'", "")
X_test['Area'] = X_test['Area'].str.replace("'", "")
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['A', 'B', 'C', 'D', 'E', 'F']])

X_train_ordinal_encoded = ordinal_encoder.fit_transform(X_train[['Area']])
X_test_ordinal_encoded = ordinal_encoder.transform(X_test[['Area']])

# VehPower auch ordinal skaliert, aber bereits numerisch
```

```python
# Überprüfung
print(X_train_ordinal_encoded[:5])
print(X_test_ordinal_encoded[:5])
```

```python
# Skalierung auf Werte zwischen 0 und 1 für neuronale Netze
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train[['VehAge','DrivAge', 'BonusMalus', 'Density_log']])
X_test_scaled = scaler.transform(X_test[['VehAge','DrivAge', 'BonusMalus', 'Density_log']])
```

```python
print(X_train_scaled.min(axis=0))
print(X_train_scaled.max(axis=0))
print(X_test_scaled.min(axis=0))
print(X_test_scaled.max(axis=0))
```

```python
# Skalierung der ordinalen Variablen (behandeln als wären intervallskaliert)
X_train_area_scaled = scaler.fit_transform(X_train_ordinal_encoded.reshape(-1, 1))
X_test_area_scaled = scaler.transform(X_test_ordinal_encoded.reshape(-1, 1))
```

```python
X_train_power_scaled = scaler.fit_transform(X_train[['VehPower']])
X_test_power_scaled = scaler.transform(X_test[['VehPower']])
```

Nicht ganz klar wofür Area und VehPower stehen. Wenn keine gleichen Abstände zwischen Kategorien angenommen werden könnten, eher wie nominal skalierte behandeln.

```python
# Daten wieder zusammenführen
X_train_trans = np.concatenate([X_train_nominal_encoded, X_train_scaled, X_train_area_scaled, X_train_power_scaled], axis=1)
X_test_trans = np.concatenate([X_test_nominal_encoded, X_test_scaled, X_test_area_scaled, X_test_power_scaled], axis=1)

print(X_train_trans.shape)
print(X_test_trans.shape)
```

```python
# Spaltennamen wieder hinzufügen
onehot_feature_names = onehot_encoder.get_feature_names_out(['VehBrand', 'VehGas', 'Region'])
scaled_feature_names = ['VehAge', 'DrivAge', 'BonusMalus', 'Density_log']  
ordinal_scaled_feature_names = ['Area', 'VehPower']
all_feature_names = np.concatenate([onehot_feature_names, scaled_feature_names, ordinal_scaled_feature_names])

X_train_trans_df = pd.DataFrame(X_train_trans, columns=all_feature_names)
X_test_trans_df = pd.DataFrame(X_test_trans, columns=all_feature_names)

print(X_train_trans_df.head())
print(X_test_trans_df.head())
```

### Baseline Modell

```python
# Baseline Modell erstellen und trainieren
rf_baseline = RandomForestRegressor()
rf_baseline.fit(X_train_trans_df, y_train)

# Vorhersagen machen
y_pred_train = rf_baseline.predict(X_train_trans_df)
y_pred_test = rf_baseline.predict(X_test_trans_df)

# RMSE berechnen
rmse_baseline_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_baseline_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print('Root Mean Squared Error für Trainingsdaten:', rmse_baseline_train)
print('Root Mean Squared Error für Testdaten:', rmse_baseline_test)

# R2-Wert für Trainings- und Testdatensatz berechnen
print(f'R2 für Trainingsdaten: {rf_baseline.score(X_train_trans_df, y_train):.3f}')
print(f'R2 für Testdaten: {rf_baseline.score(X_test_trans_df, y_test):.3f}')
```

Sehr starkes Overfitting, Modell sagt sogar schlechter vorher auf Testdaten als einfache Heuristik (negatives R2).

### Random Forest

```python
# Random Forest-Modell mit Ziel Overfitting zu verringern
rf1 = RandomForestRegressor(n_estimators=100, oob_score=True, n_jobs=-1, random_state=6876,
                            max_depth=20, min_samples_split=8, min_samples_leaf=4)

# Krossvalidierung
cv_scores = cross_val_score(rf1, X_train_trans_df, y_train, cv=5, scoring='neg_mean_squared_error')
print('CV Scores (Negative Mean Squared Error):', cv_scores)
print('Durchschnittlicher CV Score:', np.mean(cv_scores))
print('CV Scores (Root Mean Squared Error):', np.sqrt(-cv_scores))

rf1.fit(X_train_trans_df, y_train)
feature_importances = rf1.feature_importances_

print(f'R2 für Trainingsdaten: {rf1.score(X_train_trans_df, y_train):.3f}')
print(f'R2 für Testdaten: {rf1.score(X_test_trans_df, y_test):.3f}')

# Feature-Namen annehmen
feature_names = X_train_trans_df.columns

# Sortieren der Features nach ihrer Wichtigkeit
sorted_indices = np.argsort(feature_importances)[::-1]
for idx in sorted_indices:
    print(f'{feature_names[idx]}: {feature_importances[idx]}')
```

```python
# In DataFrame umwandeln
df_importances = pd.DataFrame({'Feature': feature_names, 'Importance': feature_importances})

# Sortieren der Features nach ihrer Wichtigkeit
df_importances = df_importances.sort_values(by='Importance', ascending=False)

# Plot erstellen
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=df_importances)
plt.title('Feature Importance im Random Forest Modell')
plt.xlabel('Wichtigkeit')
plt.ylabel('Feature')
plt.show()
```

```python
# Featureanzahl einschränken aufgrund von Overfitting
# Selektion aufgrund von deutlich abfallendem Beitrag nach BonusMalus
selected_features = ['Density_log', 'DrivAge', 'VehAge', 'VehPower', 'BonusMalus']

# Auswählen der Features in Datensätzen
X_train_selected = X_train_trans_df[selected_features]
X_test_selected = X_test_trans_df[selected_features]
```

Problem/Unklarheit: Ggf. zu wenig Features. Wieviel Varianz erklären sie in der AV? Was sind gängige Cutoff Punkte?

```python
# Weiteres Random Forest Modell mit ausgewählten Features
rf2 = RandomForestRegressor(n_estimators=100, oob_score=True, n_jobs=-1, random_state=6876, # Parameter konstant halten, um zu sehen, was Featureauswahl ausmacht
                            max_depth=20, min_samples_split=8, min_samples_leaf=4)

# Krossvalidierung
cv_scores = cross_val_score(rf2, X_train_selected, y_train, cv=5, scoring='neg_mean_squared_error')
print('CV Scores (Negative Mean Squared Error):', cv_scores)
print('Durchschnittlicher CV Score:', np.mean(cv_scores))
print('CV Scores (Root Mean Squared Error):', np.sqrt(-cv_scores))

rf2.fit(X_train_selected, y_train)

print(f'R2 für Trainingsdaten: {rf2.score(X_train_selected, y_train):.3f}')
print(f'R2 für Testdaten: {rf2.score(X_test_selected, y_test):.3f}')
```

Feature Selection hat kaum Verbesserung gebracht

```python
# RandomizedSearch für Hyperparameter verwenden
rf_random = RandomForestRegressor(random_state=6876)

# Parameterverteilungen definieren
param_dist = {
    'n_estimators': sp_randint(100, 500),
    'max_depth': [10, 20, 30, None],
    'min_samples_split': sp_randint(2, 11),
    'min_samples_leaf': sp_randint(1, 5),
    'max_features': ['auto', 'sqrt']
}

# Randomized Search initialisieren
random_search = RandomizedSearchCV(rf_random, param_distributions=param_dist, n_iter=100, cv=5,
                                   scoring='neg_mean_squared_error', n_jobs=-1, random_state=42)

# Randomized Search durchführen
random_search.fit(X_train_selected, y_train)

# Beste Parameter anzeigen
print('Beste Parameter:', random_search.best_params_)
```

Problem: Prozess wird unterbrochen bei erneutem Durchlauf, aber beste Parameter bereits während anderer Session ermittelt und in drittem Modell verwendet.

```python
# Random Forest-Modell mit besten Parametern
rf3 = RandomForestRegressor(n_estimators=363, oob_score=True, n_jobs=-1,random_state=6876,
                            max_depth=10, min_samples_split=8, min_samples_leaf=4, max_features='sqrt')

# Krossvalidierung
cv_scores = cross_val_score(rf3, X_train_trans_df, y_train, cv=5, scoring='neg_mean_squared_error')
print('CV Scores (Negative Mean Squared Error):', cv_scores)
print('Durchschnittlicher CV Score:', np.mean(cv_scores))
print('CV Scores (Root Mean Squared Error):', np.sqrt(-cv_scores))

rf3.fit(X_train_selected, y_train)

print(f'R2 für Trainingsdaten: {rf3.score(X_train_selected, y_train):.3f}')
print(f'R2 für Testdaten: {rf3.score(X_test_selected, y_test):.3f}')
```

```python
# Vorhersage und Bewertung
rf3_predictions = rf3.predict(X_test_selected)
mse = mean_squared_error(y_test, rf3_predictions)
print('Root Mean Squared Error:', np.sqrt(mse))

r2 = r2_score(y_test, rf3_predictions)
print('R2 auf Testdaten:', r2)
```

Modell performed immer noch sehr schlecht auf den Testdaten, aber zumindest besser als Baseline-Modell. 

### Neuronales Netz

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
# Sehr einfaches Modell als Ausgangspunkt
nn1 = Sequential()

# Eingabeschicht, 'input_dim' ist die Anzahl der Features im Datensatz
nn1.add(Dense(10, activation='relu', input_dim=X_train_trans_df.shape[1]))

# Hidden Layer
nn1.add(Dense(8, activation='relu'))

# Output-Schicht
nn1.add(Dense(1))  # Keine Aktivierungsfunktion für Regression
```

```python
nn1.compile(optimizer='adam', loss='mean_squared_error')
```

```python
nn1.fit(X_train_trans_df, y_train, epochs=50, batch_size=32, validation_split=0.2)
```

Architektur sehr simpel, Modell verbesser sich kaum über die Epochen.

```python
# Modellbewertung
loss = nn1.evaluate(X_test_trans_df, y_test)
print("Test Loss:", loss)
```

```python
# Vorhersage
nn1_predictions = nn1.predict(X_test_trans_df)
nn1_r2 = r2_score(y_test, nn1_predictions)
print("R2 auf Testdaten:", nn1_r2)
```
